## Graph Density
density of a graph tell that how much connected the nodes are in a network

```python
# Importing Libraries
import networkx as nx
import matplotlib.pyplot as plt
```

### Density for undirected graphs

$$D = \frac{E}{\frac{N(N-1)}{2}} = \frac{2E}{N(N-1)}$$

### Density for directed graphs

$$D = \frac{E}{N(N-1)}$$

```python
def density(G:nx.Graph | nx.DiGraph) -> float:
    '''
    brief : Calculates the density of a Graph G 
    param : G (nx.Graph | nx.DiGraph)
    formula : edges (Number of edges in the graph) / totalEdges (Total possible edges in a graph G)
    method : Calculate the number of nodes, edges  and apply the formula
    returns : density (float)
    '''


    # nodes (int) : stores the number of nodes in the graph G
    nodes = len(G.nodes())
    # edges (int) : stores the number of edges in the Graph G
    edges = len(G.edges())

    # calculate the density
    if (type(G) == nx.DiGraph):
        return edges/(nodes*(nodes - 1))
    return (2*edges)/(nodes*(nodes - 1))
```

```python
# Initialize an empty undirected graph
G = nx.Graph()

# Add 11 nodes to the graph
G.add_nodes_from("ABCDEFGHIJK")

# Add edges to the graph
G.add_edges_from([
    ('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('D', 'I'),
    ('E', 'F'), ('E', 'G'), ('F', 'G'), ('G', 'H'), ('H', 'I'), ('H', 'J'),
    ('I', 'K'), ('J', 'K'), ('I', 'J')
])
```

### Show Graph

```python
# pos fixes the position of the graph
pos = nx.spring_layout(G,seed=7)

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=1000, font_size=10)

# Draw edge labels (weights)
labels = nx.get_edge_attributes(G, 'weight')
plt.show()
```

```python
# Call the density function
print("Density of graph(G) : ",round(density(G),4))
```
